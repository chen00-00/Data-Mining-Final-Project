```python
import numpy as np
import pandas as pd

train = pd.read_csv("train.csv")
train.isnull().sum()
```

```python
train
```

```python
#填充缺失值
from sklearn.impute import SimpleImputer

dietary = train['Dietary Habits'].values.reshape(-1,1)
degree = train['Degree'].values.reshape(-1,1)
financial_stress = train['Financial Stress'].values.reshape(-1,1)
profession = train['Profession'].values.reshape(-1,1)
academic_pressure = train['Academic Pressure'].values.reshape(-1,1)
work_pressure = train['Work Pressure'].values.reshape(-1,1)
CGPA = train['CGPA'].values.reshape(-1,1)
study_satisfaction = train['Study Satisfaction'].values.reshape(-1,1)
job_satisfaction = train['Job Satisfaction'].values.reshape(-1,1)

imp_fre = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp_0 = SimpleImputer(missing_values = np.nan, strategy = 'constant', fill_value=0)

dietary_new = imp_fre.fit_transform(dietary)
degree_new = imp_fre.fit_transform(degree)
financial_stress_new = imp_mean.fit_transform(financial_stress)
profession_new = imp_0.fit_transform(profession)
academic_pressure_new = imp_0.fit_transform(academic_pressure)
work_pressure_new = imp_0.fit_transform(work_pressure)
CGPA_new = imp_0.fit_transform(CGPA)
study_satisfaction_new = imp_0.fit_transform(study_satisfaction)
job_satisfaction_new = imp_0.fit_transform(job_satisfaction)

train['Dietary Habits'] = pd.DataFrame(dietary_new)
train['Degree'] = pd.DataFrame(degree_new)
train['Financial Stress'] = pd.DataFrame(financial_stress_new)
train['Profession'] = pd.DataFrame(profession_new)
train['Academic Pressure'] = pd.DataFrame(academic_pressure_new)
train['Work Pressure'] = pd.DataFrame(work_pressure_new)
train['CGPA'] = pd.DataFrame(CGPA_new)
train['Study Satisfaction'] = pd.DataFrame(study_satisfaction_new)
train['Job Satisfaction'] = pd.DataFrame(job_satisfaction_new)
```

```python
test = pd.read_csv("test.csv")
print(test.isnull().sum())

train['Working Professional or Student'][train['Working Professional or Student'] == 'Working Professional'] = 1
train['Working Professional or Student'][train['Working Professional or Student'] == 'Student'] = 0
train['Have you ever had suicidal thoughts ?'][train['Have you ever had suicidal thoughts ?'] == 'Yes'] = 1
train['Have you ever had suicidal thoughts ?'][train['Have you ever had suicidal thoughts ?'] == 'No'] = 0
train['Family History of Mental Illness'][train['Family History of Mental Illness'] == 'Yes'] = 1
train['Family History of Mental Illness'][train['Family History of Mental Illness'] == 'No'] = 0

test['Working Professional or Student'][test['Working Professional or Student'] == 'Working Professional'] = 1
test['Working Professional or Student'][test['Working Professional or Student'] == 'Student'] = 0
test['Have you ever had suicidal thoughts ?'][test['Have you ever had suicidal thoughts ?'] == 'Yes'] = 1
test['Have you ever had suicidal thoughts ?'][test['Have you ever had suicidal thoughts ?'] == 'No'] = 0
test['Family History of Mental Illness'][test['Family History of Mental Illness'] == 'Yes'] = 1
test['Family History of Mental Illness'][test['Family History of Mental Illness'] == 'No'] = 0
```

```python
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier

X = pd.concat([train['Work/Study Hours'], train['Age'], train['Working Professional or Student'], train['Have you ever had suicidal thoughts ?'], train['Financial Stress'], train['Family History of Mental Illness']], axis=1)
y = train['Depression']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

depth = 10
neighbors = 6
split = 'best'
treeClass = DecisionTreeClassifier(max_depth = depth, splitter=split)
knnClass = KNeighborsClassifier(n_neighbors=neighbors, weights='uniform', metric='euclidean')
treeClass.fit(X, y)
knnClass.fit(X, y)

y_val_pred_tree = treeClass.predict(X_val)
y_val_pred_knn = knnClass.predict(X_val)
print("資料評估:")
print("depth:", depth)
print("splitter", split)
print("accuracy:", accuracy_score(y_val, y_val_pred_tree))
print("\n", classification_report(y_val, y_val_pred_tree))
print("number of neighbor", neighbors)
print("accuracy:", accuracy_score(y_val, y_val_pred_knn))
print("\n", classification_report(y_val, y_val_pred_knn))
```

```python
pred = treeClass.predict(pd.concat([test['Work/Study Hours'], test['Age'], test['Working Professional or Student'], test['Have you ever had suicidal thoughts ?'], test['Financial Stress'], test['Family History of Mental Illness']], axis=1))
test['Depression'] = pred
sub = pd.concat([test['id'], test['Depression']], axis = 1)
sub.to_csv('submission.csv',sep=',',index=False,header=True)
```
